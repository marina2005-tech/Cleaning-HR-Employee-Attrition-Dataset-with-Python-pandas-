```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_excel("HR-Employee-Attrition.xlsx")
```

```python
df.head()
```

```python
import numpy as np
df["Age Group"] = np.where(
    (df["Age"] < 20), "Bellow 20",
    np.where(
        (df["Age"] >= 20) & (df["Age"] < 30), "20-29",
        np.where(
            (df["Age"] >= 30) & (df["Age"] < 40), "30-39",
            np.where(
                (df["Age"] >= 40) & (df["Age"] < 50), "40-49",
                np.where(
                    (df["Age"] >= 50) & (df["Age"] < 60), "50-59",
                    "60+"
                )
            )
        )
    )
)
```

```python
import numpy as np

df["Distance from home range"] = np.where(
    (df["DistanceFromHome"] < 5), "0-5",
    np.where(
        (df["DistanceFromHome"] >= 5) & (df["DistanceFromHome"] <= 10), "5-10",
        np.where(
            (df["DistanceFromHome"] > 10) & (df["DistanceFromHome"] <= 25), "11-25",
            np.where(
                (df["DistanceFromHome"] >= 26) & (df["DistanceFromHome"] <= 30), "26-30",
                "Above 30"
            )
        )
    )
)
df.head()
```

```python
df["Satisfaction Label"]= np.where(
    (df["JobSatisfaction"]==1), "Very Dissatisfied",
     np.where(
    (df["JobSatisfaction"]==2), "Dissatisfied",
     np.where(
    (df["JobSatisfaction"]==3), "Neutral",
     np.where(
    (df["JobSatisfaction"]==4), "Satisfied",
     np.where(
    (df["JobSatisfaction"]==5), "Very Satisfied",
     "Unknown"
        )
      )
    )
  )
)
df.head()
```

```python
df["Performance Label"]= np.where(
    (df["PerformanceRating"]==1), "Very Low Performance",
     np.where(
    (df["PerformanceRating"]==2), "Low Performance",
     np.where(
    (df["PerformanceRating"]==3), "Good Performance",
     np.where(
    (df["PerformanceRating"]==4), "High Performance",
     np.where(
    (df["PerformanceRating"]==5), "Very High Performance",
     "Unknown"
        )
      )
    )
  )
 )
df.head()
```

```python
df.dropna(subset=["StadffID"],inplace=True)
```
